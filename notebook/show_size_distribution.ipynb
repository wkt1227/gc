```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
galaxy_sizes = np.load('../data/result/galaxy_sizes.npy')
```

```python
plt.hist(galaxy_sizes, bins=100)
```

```python
sp_sizes, el_sizes = np.split(galaxy_sizes, 2)
```

```python
plt.hist(galaxy_sizes, bins=35, label='El+Sp', range=(0, 3500), histtype='step')
plt.hist(sp_sizes, bins=35, label='Sp', range=(0, 3500), histtype='step')
plt.hist(el_sizes, bins=35, label='El', range=(0, 3500), histtype='step')

plt.legend()
plt.title('galaxy size distribution')
plt.xlabel('size(pix)')
plt.ylabel('Frequency')
```
