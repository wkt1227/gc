```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
z_list_el = np.load('../data/z_list_el.npy')
z_list_sp = np.load('../data/z_list_sp.npy')
z_list_el = z_list_el.astype(np.float)
z_list_sp = z_list_sp.astype(np.float)
z_list = np.append(z_list_sp, z_list_el)
```

```python
plt.hist(z_list, bins=36, range=(0, 0.36), alpha=0.4, label='El+Sp', )
plt.hist(z_list_sp, bins=36, range=(0, 0.36), alpha=0.4, label='Sp', )
plt.hist(z_list_el, bins=36, range=(0, 0.36), alpha=0.4, label='El', )

plt.legend()
plt.title('z distribution')
plt.xlabel('z')
plt.ylabel('Frequency')
```
